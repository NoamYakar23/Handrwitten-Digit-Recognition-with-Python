```python
import numpy as np
import mnist 
import matplotlib.pyplot as plt

from PIL import Image as im
from IPython.display import display

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
```

```python
train_imgs = mnist.train_images()
train_lbls = mnist.train_labels()

test_imgs =  mnist.test_images()
test_lbls =  mnist.test_labels()
```

**Normalize Our Images**

```python
train_imgs = (train_imgs / 255) - 0.5
test_imgs = (test_imgs / 255) - 0.5
```

**Flattened Images**

```python
train_imgs = train_imgs.reshape((-1,784))
test_imgs = test_imgs.reshape((-1,784))
```

**Built Model**

```python
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(784,)))
model.add(Dense(64, activation='relu'))
model.add(Dense(10, activation = 'softmax'))
```

**Compiled & Fit The Model**

```python
model.compile(
optimizer = 'adam',
    loss = 'categorical_crossentropy',
    metrics=['accuracy']
)

model.fit(train_imgs, to_categorical(train_lbls), epochs = 10, batch_size=128)

```

```python
model.evaluate(test_imgs, to_categorical(test_lbls))
model.save_weights('model.h5')

```
